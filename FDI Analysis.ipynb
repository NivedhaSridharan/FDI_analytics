```python
Yrs = ['2000-01', '2001-02', '2002-03', '2003-04', '2004-05','2005-06', '2006-07', '2007-08', '2008-09',
          '2009-10', '2010-11','2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17']
Sect= ['Sector']
```

```python

# Exchange rate 2000-2017
ExchangeRt_Avg = [45.68,47.69,48.39,45.95,44.93,44.27,45.25,40.26,45.99,47.44,45.56,47.92,54.40,
                  60.50,61.14,65.47,67.07]
```

```python
spdUSD = data.copy()
```

```python
# here  df is FDI_spd column taken as year rates as exchange rate
def Convers_Usdinr(df,colm,rts):
    for ele in colm:
        df[ele] = df[ele] * rts[colm.index(ele)]/10
    return df    

spdINR = Convers_Usdinr(data, Yrs, ExchangeRt_Avg)
spdINR 
```

```python
#amount in Indian Rupees
spdINR.style.set_caption("FDI Amount in INR Crores").format(precision=2)
spdINR
```

##### melt() is used to analyse and visualize the data better.

```python
melt = pd.melt(spdUSD, id_vars = Sect, value_vars = Yrs, var_name='Year', value_name='FDI(USD Million)',ignore_index=True)
melt
```

```python
melt=round(melt,2)
melt
```

```python
melt_01 = pd.melt(spdINR, id_vars = Sect, value_vars = Yrs, var_name='Year',value_name='FDI(INR Crores)',ignore_index=True)
melt_01
```

```python
# Rounding off upto two decimal point
melt_01=round(melt_01,2)
melt_01
```

```python
# Merging the FDI(USD Million) column of melt Dataframe into melt_01 Dataframe
Fdi_Merge= melt_01.merge(melt,how='left')
Fdi_Merge
```

```python
#Sorting the Sector and Year columns
Fdi_Sort = Fdi_Merge.sort_values(['Sector','Year'], ignore_index=True)
Fdi_Sort
```

##### Finding the Statistics of th column Sector

```python
print("\nStatistics for Sectors\n",'-'*65, sep='')
print(pd.DataFrame(Fdi_Sort.groupby('Sector').describe().loc[:,:]).transpose())
```

```python
#Grouping by the Sector column to find Total FDI Inflow per Sector from 2000-01 to 2016-17  
FDI_Sectws = Fdi_Sort.groupby('Sector').sum()
FDI_Sectws.sort_values(by='FDI(USD Million)',ascending=False)
```

##### Bar graph representing the Amount in USD 

```python
FDI_Sectws.plot(kind='bar',y='FDI(USD Million)',figsize = (15,8))
```

##### Bar gragh representing the Amount in INR

```python
FDI_Sectws.plot(kind='bar',y='FDI(INR Crores)',figsize = (15,8))
```

```python
#Creating Dataframe  
melt_02 = melt_01[['Year', 'FDI(INR Crores)']]
melt_02= round(melt_02.groupby('Year').sum(),2)
melt_02
```

###### Percentage increase of FDI in contrast to previous years

```python
melt_02['% growth over previous year'] = round(melt_02.pct_change()*100,2)
melt_02.fillna('-')
```

From the above data, we could report that over a time period between 2000-2017 there have been some significant increase and also some significant decrease happened in the FDI inflows to the country.
We can say that the year 2002-04 has the lowest inflows compared to the year 2001-02. 
The year 2006-07 has more inflows than the previous years between 2000-06.
In 2014-2015, it increased its inflows after the downfall. 
The last two years, 2015-17 the inflows increased enormously compared to the previous years from 2000-14.

##### Line graph representing the INR in crores

```python
melt_02.plot.line(y='FDI(INR Crores)',figsize=(35,15))
plt.ylabel('FDI(INR Crores)')
plt.show()
```

###### Identifying the highest FDI

```python
Sect_HighFdi = FDI_Sectws.nlargest(10,['FDI(INR Crores)'])
Fdiin_Tot = round(melt_01['FDI(INR Crores)'].sum(),2)
Sum = Sect_HighFdi['FDI(INR Crores)'].sum()
Sect_HighFdi['In %age'] = round(Sect_HighFdi['FDI(INR Crores)']/Sum*100,2) 
Sect_HighFdi['%age to Total Inflows'] = round((Sect_HighFdi['FDI(INR Crores)']/Fdiin_Tot)*100,2)
Sect_HighFdi
```

###### Identifyin the lowest FDI

```python
Sect_LowFdi = FDI_Sectws.nsmallest(10,['FDI(INR Crores)'])
Sum = Sect_LowFdi['FDI(INR Crores)'].sum()
Sect_LowFdi['In %age'] = round(Sect_LowFdi['FDI(INR Crores)']/Sum*100,2)
Sect_LowFdi['%age to Total Inflows'] = round((Sect_LowFdi['FDI(INR Crores)']/Fdiin_Tot)*100,3)
Sect_LowFdi
```

###### Bar graph for highest FDI

```python
plt.figure(figsize=(20,10))
plt.barh(Sect_HighFdi.index,Sect_HighFdi['FDI(INR Crores)'])
plt.xlabel('FDI(INR Crores)')
plt.ylabel('Sectors')
plt.show()
```

###### Bar graph for lowest FDI

```python
plt.figure(figsize=(20,10))
plt.barh(Sect_LowFdi.index,Sect_LowFdi['FDI(INR Crores)'])
plt.xlabel('FDI(INR Crores)')
plt.ylabel('Bottom Sectors')
plt.show()
```
